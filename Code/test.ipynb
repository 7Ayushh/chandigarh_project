```python
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
X_enc = df[categorical_columns]
X_enc = pd.get_dummies(X_enc)
X_enc
```

```python
pd.set_option('display.max_columns', None)
X_enc.columns = X_enc.columns.str.replace('[\[\]<>]', '', regex=True)  # Remove problematic characters
X_enc.columns = X_enc.columns.str.strip()  # Remove leading/trailing spaces
print(X_enc.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, random_state= 100)
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
from xgboost import XGBRegressor
xg_reg = XGBRegressor()
```

```python
xg_reg.fit(X_train, y_train, verbose=True)
```

```python
y_pred = xg_reg.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("MSE: %0.2f"% (mse))
print("RMSE: %0.2f" % (np.sqrt(mse)))
```

```python
from sklearn.model_selection import cross_val_score
model = XGBRegressor(objective="reg:squarederror")
scores = cross_val_score(model, X_enc, y, scoring="neg_mean_squared_error", cv=10)
print(scores)
```
